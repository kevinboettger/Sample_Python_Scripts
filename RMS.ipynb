```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#include audio files
debussy_file = "audio_debussy.wav"
redhot_file = "audio_redhot.wav"
duke_file = "audio_duke.wav"
```

```python
#load the files with librosa

debussy, sr = librosa.load(debussy_file)
redhot, sr = librosa.load(redhot_file)
duke, sr = librosa.load(duke_file)
```

```python
#extract rms energy with librosa

FRAME_SIZE = 1024
HOP_LENGTH = 512


rms_debussy = librosa.feature.rms(debussy,frame_length = FRAME_SIZE, hop_length = HOP_LENGTH)[0]
rms_redhot = librosa.feature.rms(redhot,frame_length = FRAME_SIZE, hop_length = HOP_LENGTH)[0]
rms_duke = librosa.feature.rms(duke,frame_length = FRAME_SIZE, hop_length = HOP_LENGTH)[0]
```

```python
#plot the rms energy for the 3 audio files

frames = range(0,rms_debussy.size)
t = librosa.frames_to_time(frames, hop_length=HOP_LENGTH)

plt.figure(figsize=(15,17))

plt.subplot(3,1,1)
librosa.display.waveplot(debussy)
plt.plot(t,rms_debussy,color ='r')
plt.title('Debussy')
plt.ylim(-1,1)
plt.xlabel('Time')
plt.ylabel('Amplitude')


plt.subplot(3,1,2)
librosa.display.waveplot(redhot)
plt.plot(t,rms_redhot,color ='r')
plt.title('RHCP')
plt.ylim(-1,1)
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(3,1,3)
librosa.display.waveplot(duke)
plt.plot(t,rms_duke,color ='r')
plt.title('Duke')
plt.ylim(-1,1)
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.show()
```

```python
def rms(signal, frame_length, hop_length):
    rms = []
    
    for i in range(0,len(signal),hop_length):
        rms_current_frame = np.sqrt(np.sum(signal[i:i+frame_length]**2)/frame_length)
        rms.append(rms_current_frame)
        
    return np.array(rms)
        
```

```python
rms1_debussy = rms(debussy,frame_length = FRAME_SIZE, hop_length = HOP_LENGTH)
rms1_redhot = rms(redhot,frame_length = FRAME_SIZE, hop_length = HOP_LENGTH)
rms1_duke = rms(duke,frame_length = FRAME_SIZE, hop_length = HOP_LENGTH)
```

```python
plt.figure(figsize=(15,17))

plt.subplot(3,1,1)
librosa.display.waveplot(debussy)
plt.plot(t,rms_debussy,color ='r')
plt.plot(t,rms1_debussy, color='y')
plt.title('Debussy')
plt.ylim(-1,1)
plt.xlabel('Time')
plt.ylabel('Amplitude')


plt.subplot(3,1,2)
librosa.display.waveplot(redhot)
plt.plot(t,rms_redhot,color ='r')
plt.plot(t,rms1_redhot,color = 'y')
plt.title('RHCP')
plt.ylim(-1,1)
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(3,1,3)
librosa.display.waveplot(duke)
plt.plot(t,rms_duke,color ='r')
plt.plot(t,rms1_duke,color = 'y')
plt.title('Duke')
plt.ylim(-1,1)
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.show()
```

```python
#zero-crossing rate

zcr_debussy = librosa.feature.zero_crossing_rate(debussy,frame_length = FRAME_SIZE, hop_length = HOP_LENGTH)[0]
zcr_redhot = librosa.feature.zero_crossing_rate(redhot,frame_length = FRAME_SIZE, hop_length = HOP_LENGTH)[0]
zcr_duke = librosa.feature.zero_crossing_rate(duke,frame_length = FRAME_SIZE, hop_length = HOP_LENGTH)[0]
```

```python
#visualize the zcr for all audio

plt.figure(figsize=(15,17))
plt.plot(t,zcr_debussy * FRAME_SIZE,color ='r')
plt.plot(t,zcr_redhot * FRAME_SIZE,color ='b')
plt.plot(t,zcr_duke * FRAME_SIZE,color ='y')

plt.ylim(0,500)



plt.show()
```
